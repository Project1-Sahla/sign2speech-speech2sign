## Create a class called none


to train mediapipe geture recoginition model we need that our dataset has a class called none that has getures that doesn't belong to any class. we created this none class by taking images from a dataset called rps_data_sample and taking all the none images of this dataset and took some images from the other classes and create a none folder for our data

```python
# create a null folder
import shutil
import os
import random

# Paths
source_folder = '/content/drive/MyDrive/project1/rps_data_sample'
target_folder = '/content/drive/MyDrive/project1/RGB_ArSL_dataset/none'

# Ensure the target directory exists (create manually in Google Drive if needed)
os.makedirs(target_folder, exist_ok=True)

# Step 1: Copy all 125 images from the 'none' folder
none_folder = os.path.join(source_folder, 'none')
none_images = os.listdir(none_folder)[:125]
for image in none_images:
    shutil.copy(os.path.join(none_folder, image), target_folder)

# Step 2: Copy 35 random images from each of the 'paper', 'rock', and 'scissors' folders
for folder in ['paper', 'rock', 'scissors']:
    folder_path = os.path.join(source_folder, folder)
    images = os.listdir(folder_path)
    selected_images = random.sample(images, 35)  # Select 35 random images
    for image in selected_images:
        shutil.copy(os.path.join(folder_path, image), target_folder)

print("Images copied successfully to the target folder.")
```

## creating a sample folder with 5 images for each class

```python
# create a sample folder
import os
import shutil
import random

def create_sample_folder(original_folder, sample_folder, image_limit=5):
    # Ensure the sample folder exists
    if not os.path.exists(sample_folder):
        os.makedirs(sample_folder)

    # Traverse the original folder structure
    for root, dirs, files in os.walk(original_folder):
        # Relative path from original folder to the current folder
        relative_path = os.path.relpath(root, original_folder)
        # Destination path in the sample folder
        dest_dir = os.path.join(sample_folder, relative_path)

        # Create the directory structure in the sample folder
        if not os.path.exists(dest_dir):
            os.makedirs(dest_dir)

        # Filter image files (can adjust extensions if needed)
        image_files = [f for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]

        # Limit the number of images to the specified limit (default is 5)
        sample_images = random.sample(image_files, min(len(image_files), image_limit))

        # Copy the selected images to the sample folder
        for image in sample_images:
            src_path = os.path.join(root, image)
            dest_path = os.path.join(dest_dir, image)
            shutil.copy(src_path, dest_path)

    print(f"Sample folder created at: {sample_folder}")

# Define paths for the original folder and the sample folder
original_folder = "/content/drive/MyDrive/project1/RGB_ArSL_dataset"
sample_folder = "/content/drive/MyDrive/project1/RGB_ArSL_dataset_sample"

# Run the function to create a sample folder
create_sample_folder(original_folder, sample_folder)
```
